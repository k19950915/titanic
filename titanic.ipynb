```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
train_df
```

```python
train_df.isna().sum()
```

```python
test_df.isna().sum()
```

```python
med = train_df.median()
train_df = train_df.fillna(med)
train_df.isna().sum()
```

```python
test_df = test_df.fillna(med)
test_df.isna().sum()
```

```python
the_most = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(the_most)
train_df.isna().sum()
```

```python
test_df["Embarked"] = test_df["Embarked"].fillna(the_most)
test_df.isna().sum()
```

```python
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
dummy = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
dummy = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
s = "Braund, Mr. Owen Harris"
s.split(",")[-1].split(".")[0].replace(" ", "")
```

```python
def namecut(s):
    return s.split(",")[-1].split(".")[0].replace(" ", "")
mid = train_df["Name"].apply(namecut)
pd.crosstab(train_df["Survived"], mid)
```

```python
def namecut(s):
    reserved = ["Master", "Miss", "Mr", "Mrs"]
    m = s.split(",")[-1].split(".")[0].replace(" ", "")
    if m in reserved:
        return m
    else:
        return "X"
trans = train_df["Name"].apply(namecut)
dummy = pd.get_dummies(trans)
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
trans = test_df["Name"].apply(namecut)
dummy = pd.get_dummies(trans)
test_df = pd.concat([test_df, dummy], axis=1)
test_df
```

```python
train_df.columns
```

```python
train = train_df.drop(['PassengerId', 'Name', 
                       'Sex', 'Ticket', 
                       'Cabin', 'Embarked', 
                       'X'], axis=1)
x_train = train.drop(["Survived"], axis=1)
y_train = train["Survived"]
x_train
```

```python
test = test_df.drop(['Name', 'Sex', 'Ticket', 
                       'Cabin', 'Embarked', 'X'], 
                     axis=1)
x_test = test.drop(["PassengerId"], axis=1)
pid = test["PassengerId"]
x_test
```

```python
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(14, 10))
sns.heatmap(train.corr(), annot=True, cmap="RdBu_r")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
p = {
    "n_estimators":range(15, 25),
    "max_depth":range(5, 10)
}
cv = GridSearchCV(clf, param_grid=p, cv=10)
cv.fit(x_train, y_train)
cv.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
clf = RandomForestClassifier(n_estimators=26, max_depth=6)
np.average(cross_val_score(clf, x_train, y_train, cv=10))
```

```python
clf = RandomForestClassifier(n_estimators=26, max_depth=7)
clf.fit(x_train, y_train)
```

```python
pre = clf.predict(x_test)
result = pd.DataFrame()
result["PassengerId"] = pid
result["Survived"] = pre
result.to_csv("titanic.csv", encoding="utf-8", index=False)
```
